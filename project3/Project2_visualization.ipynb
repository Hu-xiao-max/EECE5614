# Maze Visualization

```python
"""
Visualization code to display found State Values and the optimum policy on the maze. 
"""
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
      

Env = np.zeros([20, 20])   # The matrix which describe the environment
# regular = 0    wall = 1    oil = 2     bump = 3    start = 4    end = 5
Env[:, [0]] = np.ones([20, 1])
Env[:, [19]] = np.ones([20, 1])
Env[0, :] = np.ones([1, 20])
Env[19, :] = np.ones([1, 20])
Env[1, 11] = 3; Env[1, 12] = 3
Env[2, 1:4] = [3, 3, 3]; Env[2, 5] = 1; Env[2, 8] = 2; Env[2, 16] = 2
Env[3, 5] = 1; Env[3, 13] = 5
Env[4, 2] = 2; Env[4, 3:17] = np.ones([1, 14])
Env[5, 1] = 3; Env[5, 3] = 1; Env[5, 6] = 2; Env[5, 9] = 3; Env[5, 17] = 3
Env[6, 3] = 1; Env[6, 6] = 1; Env[6, 9] = 1; Env[6, 15] = 1; Env[6, 17] = 3
Env[7, 2] = 3; Env[7, 3] = 1; Env[7, 6] = 1; Env[7, 9] = 1; Env[7, 10:12] = [3, 3]; Env[7, 12:16] = np.ones([1, 4]); Env[7, 17] = 3
Env[8, 6] = 1; Env[8, 9] = 1; Env[8, 15] = 1; Env[8, 17] = 3
Env[9, 6] = 1; Env[9, 9] = 1; Env[9, 15] = 1
Env[10, 1:5] = np.ones([1, 4]); Env[10, 6] = 1; Env[10, 9:11] = [1, 1]; Env[10, 15] = 1; Env[10, 18] = 2
Env[11, 6] = 1; Env[11, 10] = 1; Env[11, 13] = 1; Env[11, 15:18] = [1, 1, 1]
Env[12, 3:8] = np.ones([1, 5]); Env[12, 10] = 1; Env[12, 11:13] = [3, 3]; Env[12, 13] = 1; Env[12, 17] = 1
Env[13, 7] = 1; Env[13, 10] = 1; Env[13, 13] = 1; Env[13, 17] = 1
Env[14, 1:3] = [3, 3]; Env[14, 7] = 1; Env[14, 10] = 1; Env[14, 13] = 1
Env[15, 4] = 4; Env[15, 7] = 1; Env[15, 10] = 2; Env[15, 13:17] = [1, 1, 1, 1]; Env[15, 17:19] = [3, 3]
Env[16, 7] = 3; Env[16, 10] = 2
Env[17, 1:3] = [1, 1]; Env[17, 7:13] = [1, 1, 1, 1, 1, 1]; Env[17, 14] = 2; Env[17, 17] = 2
Env[18, 7] = 2

# Define colors
colors = {
    0: [1, 1, 1],        # White
    1: [0, 0, 0],        # Black
    2: [0.55, 0, 0],     # Light Brown
    3: [0.96, 0.8, 0.6], # Dark Red
    4: [0, 0, 1],        # Green
    5: [0, 1, 0]         # Blue
}


rgb_maze = np.zeros((Env.shape[0], Env.shape[1], 3))
for i in range(Env.shape[0]):
    for j in range(Env.shape[1]):
        rgb_maze[i, j] = colors.get(Env[i, j], [1, 1, 1])


plt.figure(figsize=(15, 10))
sns.heatmap(Env,fmt="",  cmap=sns.color_palette([colors[i] for i in range(6)]), cbar=False,annot=False, linewidths=0.5, linecolor='black')
plt.axis('off')
plt.title('Maze Problem')
plt.show()


```

# Plotting path on the maze

```python
plt.figure(figsize=(15, 10))
sns.heatmap(Env,fmt="",  cmap=sns.color_palette([colors[i] for i in range(6)]), cbar=False,annot=False, linewidths=0.5, linecolor='black')
plt.axis('off')
plt.title('Maze Problem - An example path')


# This is just an example, you may want to keep pi* coordinates and actions in a different way
path = [((3,3),'right'), ((3,4),'up'), ((2,4),'up'), ((1,4),'right'), \
        ((1,5),'right'), ((1,6),'down'), ((2,6),'right'),((2,7),'right')]


for state_cr, direction in path:
    r = state_cr[0] # x_coordinate
    c = state_cr[1] # y_coordinate

    if direction == 'right':
        plt.arrow(c + 0.5, r + 0.5, 0.8, 0, width=0.04, color='black')   # Right
    if direction == 'left':
        plt.arrow(c + 0.5, r + 0.5, -0.8, 0, width=0.04, color='black')  # Left
    if direction == 'up':
        plt.arrow(c + 0.5, r + 0.5, 0, -0.8, width=0.04, color='black')  # Up
    if direction == 'down':
        plt.arrow(c + 0.5, r + 0.5, 0, 0.8, width=0.04, color='black')  # Down



plt.show()
```

# Plot policy on the maze

```python
plt.figure(figsize=(15, 10))
sns.heatmap(Env,fmt="",  cmap=sns.color_palette([colors[i] for i in range(6)]), cbar=False,annot=False, linewidths=0.5, linecolor='black')
plt.axis('off')
plt.title('Maze Problem - A random policy')

directions = ["up", "right", "down", "left"]
random_policy = np.random.choice(directions, size=248) # A vector with size 248 which each element is a direction

for i in range(20):
    for j in range(20):
        if State_Matrix[i, j] != 0:
            if random_policy[State_Matrix[i, j]-1] == "up":
                    plt.arrow(j+0.5, i+0.85, 0, -0.5, width=0.04, color='black')  # Up
            if random_policy[State_Matrix[i, j]-1] == "right":
                    plt.arrow(j+0.15, i+0.5, 0.5, 0, width=0.04, color='black')  # Right
            if random_policy[State_Matrix[i, j]-1] == "down":
                    plt.arrow(j+0.5, i+0.15, 0, 0.50, width=0.04, color='black')  # Down
            if random_policy[State_Matrix[i, j]-1] == "left":
                    plt.arrow(j+0.85, i+0.5, -0.5, 0, width=0.04, color='black')  # Left
```
